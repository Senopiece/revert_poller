```python
import pandas as pd
import matplotlib.pyplot as plt

# Function to plot the CSV data
def plot_financial_metrics(filename):
    # Read the CSV data into a DataFrame
    data = pd.read_csv(filename, parse_dates=["Time"])
    
    # Plot each column
    plt.figure(figsize=(12, 8))

    # Pooled Assets
    plt.plot(data["Time"], data["Pooled Assets $"], label="Pooled Assets $")
    # Total PnL
    plt.plot(data["Time"], data["Total PnL $"], label="Total PnL $")
    # Total APR %
    plt.plot(data["Time"], data["Total APR %"], label="Total APR %")
    # Fee APR %
    plt.plot(data["Time"], data["Fee APR %"], label="Fee APR %")

    # Customize the plot
    plt.title("Financial Metrics Over Time")
    plt.xlabel("Time")
    plt.ylabel("Values")
    plt.legend()
    plt.grid()
    plt.xticks(rotation=45)
    plt.tight_layout()

    # Display the plot
    plt.show()

# Specify the CSV filename
csv_filename = "charts/NMT-USDT.csv"  # Replace with pair you're interested in
plot_financial_metrics(csv_filename)
```
